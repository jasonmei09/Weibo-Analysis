# 反扒教学

## 常见反扒技术

* **User-Agent**: 网站通过验证接收到的 User-Agent 信息中是否包含 python 字样来拒绝访问呢
* **Referer**: 表示请求从哪发出的。网站可以指定从某个地址跳转进来，对于其他的地址一律禁止访问
* **authorization**: 网站可以禁止没有用户ID的访问，我们那可以在header中添加client-ID来解决这个问题
* **添加延迟**:通过用time.sleep(1)是脚本休息1秒的方式来防止反爬虫

````python
import requests

# 关键字
url = 'https://image.baidu.com/search/index?tn=baiduimage&ipn=r&ct=201326592&cl=2&lm=-1&st=-1&fm=result&fr=&sf=1&fmq=1539490550614_R&pv=&ic=0&nc=1&z=&se=1&showtab=0&fb=0&width=&height=&face=0&istype=2&ie=utf-8'
keyword = "刘德华"

try:
    kv = {"word":keyword}
    r = requests.get(url, params = kv)
    print(r.request.url)
    r.raise_for_status()
    print(len(r.text))
    #print(r.text)
except:
    print("爬取失败")
````

````python
import requests
import os

url = "http://image.nationalgeographic.com.cn/2015/0121/20150121033625957.jpg"
root = "Images/"
path = root + url.split('/')[-1]

try:
    if not os.path.exists(root):
        os.mkdir(root)
    if not os.path.exists(path):
        r = requests.get(url)
        with open(path,'wb') as f:
            f.write(r.content)
            f.close()
            print("文件保存成功")
    else :
        print("文件已存在")
except:
    print("爬取失败")
````

### 基本概念
* HTTP协议是用于**从万维网服务器传输超文本到客户端（本地浏览器）的传送协议**
* 基于 TCP/IP 协议来传递数据
* 客户端**通过URL向服务器发送所有请求**
* 无连接（服务器限制每次连接只处理一个请求，处理完后，即刻断开连接）
* 媒体独立（任何类型的数据都可通过HTTP发送）
* 无状态（服务器对于事务处理没有记忆能力）

> ### URL
* http://www.aspxfans.com:8080/news/index.asp?boardID=5&ID=24618&page=1#name
* "http:": 说明网页用的是HTTP协议
* "www.aspxfans.com": 域名
* ":8080": 端口
* "/news/": 虚拟目录部分
* "index.asp": 文件名部分, 位于最后一个 "/" 和 "?" 之间
* "boardID=5&ID=24618&page=1": 参数（搜索）部分, 位于 "?" 和 "#" 之间
* "name": M锚部分, 从 "#" 开始到最后

### HTTP请求
1. get: 请求指定的页面信息
2. head: 请求指定的页面的报头
3. post: 向指定资源提交数据进行处理请求（例如上传文件，提交表单）。post可以更改现有资源，以及建立新的资源
4. put: 取代服务器内指定文档的内容
5. delete: 请求服务器删除指定页面
6. options: 运行客户端查看服务器的性能
